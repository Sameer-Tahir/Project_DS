```python
      # Import libraries
import pandas as pd   # data processing
import numpy as np    # linear algebra
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/features of hpc and kerner not null.csv", header=None)
```

```python
df
```

```python
# Step 1: Extract feature names
kernel_features = df.iloc[0].dropna().tolist()
hpc_features_raw = df.iloc[1].dropna().tolist()

# Step 2: Normalize HPC feature names
def normalize_hpc_name(name):
    return name.replace(":", "_")

hpc_features = [normalize_hpc_name(f) for f in hpc_features_raw]

# Step 3: Convert to sets for comparison
kernel_set = set(kernel_features)
hpc_set = set(hpc_features)

# Step 4: Compute intersections and differences
shared_features = sorted(kernel_set & hpc_set)
kernel_only = sorted(kernel_set - hpc_set)
hpc_only = sorted(hpc_set - kernel_set)

# Step 5: Print summary
print(f"üîÅ Shared features: {len(shared_features)}")
print(f"üß† Kernel-only features: {len(kernel_only)}")
print(f"‚öôÔ∏è HPC-only features: {len(hpc_only)}")

# Optional: Save to dictionary for further use
feature_summary = {
    "shared": shared_features,
    "kernel_only": kernel_only,
    "hpc_only": hpc_only
}
```

```python
# Step 1: Extract and normalize
kernel_features = df.iloc[0].dropna().tolist()
hpc_raw = df.iloc[1].dropna().tolist()
hpc_features = [f.replace(":", "_") for f in hpc_raw]

# Step 2: Convert to sets
kernel_set = set(kernel_features)
hpc_set = set(hpc_features)

# Step 3: Compute groups
shared = sorted(kernel_set & hpc_set)
kernel_only = sorted(kernel_set - hpc_set)
hpc_only = sorted(hpc_set - kernel_set)

# Step 4: Create side-by-side table for shared features
shared_df = pd.DataFrame({
    "Kernel Feature": shared,
    "HPC Feature": shared
})

# Step 5: Create separate tables for unique features
kernel_only_df = pd.DataFrame({"Kernel-Only Feature": kernel_only})
hpc_only_df = pd.DataFrame({"HPC-Only Feature": hpc_only})

# Step 6: Display or export
print("‚úÖ Shared Features (Side-by-Side):")
print(shared_df)

print("\nüß† Kernel-Only Features:")
print(kernel_only_df)

print("\n‚öôÔ∏è HPC-Only Features:")
print(hpc_only_df)
```
