```python
rows_with_nans = kernel_events[kernel_events.isnull().any(axis=1)]
print(rows_with_nans)
```

```python
# Find the position of the "interface" column
interface_loc = kernel_events.columns.get_loc("interface")
print("Interface column index:", interface_loc)

# Check columns after "interface"
print("Columns after interface:", kernel_events.columns[interface_loc+1:])
```

```python
# Drop unwanted columns (e.g., Unnamed: 911 → Unnamed: 914)
kernel_events = kernel_events.drop(kernel_events.columns[interface_loc+1:], axis=1)

print("Shape after dropping:", kernel_events.shape)
```

```python
columns_with_nans = kernel_events.columns[kernel_events.isnull().any()]
print(columns_with_nans)
```

```python
for idx in rows_with_nans.index:
    missing_cols = rows_with_nans.loc[idx].index[rows_with_nans.loc[idx].isnull()]
    print(f"Row {idx} has NaN in columns: {list(missing_cols)}")
```

```python
duplicate_cols = kernel_events.columns[kernel_events.columns.duplicated()]
print("Duplicate Columns:", duplicate_cols.tolist())
```

```python
duplicate_rows = kernel_events[kernel_events.duplicated()]
print("Duplicate Rows:", duplicate_rows)
```

```python
print("Shape after cleaning:", kernel_events.shape)
```

```python
constant_cols = [col for col in kernel_events.columns if kernel_events[col].nunique() <= 1]
print("Constant Columns:", constant_cols)
```

```python
# Select only numeric columns
numeric_df = kernel_events.select_dtypes(include=['number', 'bool'])

# Compute correlation
corr_matrix = numeric_df.corr().abs()

# Get upper triangle of the correlation matrix
upper_tri = corr_matrix.where(~np.tril(np.ones(corr_matrix.shape)).astype(bool))

# Find highly correlated features (> 0.95 for example)
high_corr = [col for col in upper_tri.columns if any(upper_tri[col] < 0.10)]
print("Highly correlated columns to consider dropping:", high_corr)
```

```python
# Extract the unique values of 'interface'
unique_vals = kernel_events['interface'].unique()

# Print all unique values with type and repr (to reveal hidden spaces/special chars)
for val in unique_vals:
    print(f"Value: {val} | Type: {type(val)} | Raw: {repr(val)}")

# Specifically compare only the 'iso15118' variants
iso_vals = [val for val in unique_vals if "iso15118" in str(val).lower()]

print("\n--- Comparing iso15118 variants ---")
for i, v in enumerate(iso_vals):
    print(f"iso_val[{i}]: {repr(v)} (Type: {type(v)})")

# Check if they are truly equal
if len(iso_vals) == 2:
    print("Are they equal (==)?", iso_vals[0] == iso_vals[1])
    print("Length of each string:", len(iso_vals[0]), len(iso_vals[1]))
    print("Char codes:")
    print([ord(c) for c in iso_vals[0]])
    print([ord(c) for c in iso_vals[1]])
    print(kernel_events.shape)
```
