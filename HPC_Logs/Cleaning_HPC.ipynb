```python
constant_cols = [col for col in hpc_events.columns if hpc_events[col].nunique() <= 1]
print("Constant Columns:", constant_cols)
```

```python
# Select only numeric columns
numeric_df = hpc_events.select_dtypes(include=['number', 'bool'])

# Compute correlation
corr_matrix = numeric_df.corr().abs()

# Get upper triangle of the correlation matrix
upper_tri = corr_matrix.where(~np.tril(np.ones(corr_matrix.shape)).astype(bool))

# Find highly correlated features (> 0.95 for example)
high_corr = [col for col in upper_tri.columns if any(upper_tri[col] < 0.10)]
print("Highly correlated columns to consider dropping:", high_corr)
```
